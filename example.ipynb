```python
import seaborn as sns
import matplotlib.pyplot as plt

from dunits import read_units, read_alignments, count_matrix, proba_phone_code, units_quality
```

```python
units = read_units("dev-clean.km", "dev-clean.tsv") | read_units("dev-other.km", "dev-other.tsv")
phones = read_alignments("dev-clean.align") | read_alignments("dev-other.align")
count, phone_order = count_matrix(phones, units)
```

```python
units_quality(count)
plt.figure(figsize=(8, 8))
sns.heatmap(proba_phone_code(count)[0], cmap="Blues", yticklabels=phone_order, xticklabels=False)
plt.show()
```
